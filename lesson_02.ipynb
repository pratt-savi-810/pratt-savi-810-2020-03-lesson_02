````python
import geopandas as gpd

nyc_community_districts_geojson = 'https://data.cityofnewyork.us/api/geospatial/yfnk-k7r4?method=export&format=GeoJSON'
nyc_community_districts = gpd.read_file(nyc_community_districts_geojson)
nyc_wifi_geojson = "https://data.cityofnewyork.us/resource/yjub-udmw.geojson"
favelas_geojson = nyc_community_districts.to_crs(epsg='4326').to_json()
````

````python
import folium

m = folium.Map(location=[45.5236, -122.6750])

def style_favelas_geojson(feature):
    return {
        'fillOpacity': 0.8,
        'weight': 4,
        'fillColor': '#66ff99',
        'color': '#001a09',
    }

folium.GeoJson(
    nyc_wifi_geojson, 
    name='Census Tracts Example', 
    style_function=style_favelas_geojson,
).add_to(m)

folium.LayerControl().add_to(m)

# m
````

````python

m = folium.Map(
    location=[45.372, -121.6972],
    zoom_start=12,
    tiles='Stamen Terrain'
)

tooltip = 'Click me!'

folium.Marker([45.3288, -121.6625], popup='<i>Mt. Hood Meadows</i>', tooltip=tooltip).add_to(m)
folium.Marker([45.3311, -121.7113], popup='<b>Timberline Lodge</b>', tooltip=tooltip).add_to(m)

m
````
